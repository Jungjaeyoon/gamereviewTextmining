# Review data scrap from gamelisturl.csv

### 실제 리뷰 데이터를 url에서 받아 저장하기

```python
from bs4 import BeautifulSoup as bsoup
import requests 
import pandas as pd
import re
```

```python
gamelist_data= pd.read_csv(r'C:\Users\jaeyoon\GoogleDrive\re\game\gamelist_data.csv')
gamelist_data = gamelist_data.drop('Unnamed: 0',1)
```

```python
gamelist_data = gamelist_data.iloc[0:3]
```

```python
gamelist_data
```

#### critic, user review를 게임마다 추출하는 url 기본형 지정

```python
basegamepageurl = 'http://www.metacritic.com/'
basegamepageurl_critic = 'http://www.metacritic.com/%s/critic-reviews'
basegamepageurl_user = 'http://www.metacritic.com/%s/user-reviews?sort-by=most-helpful&num_items=100'
urlheader = {'User-Agent' :'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:54.0) Gecko/20100101 Firefox/54.0' }
```

```python
review_data_list = []
```

#### 이제 각 게임에 대한 리뷰를 받습니다. 각 타이틀 별로 존재하는 리뷰를 받는 코드 입니다. 앞서 만든 데이터 프레임에 속하는 데이터를 가져와 데이터 프레임에 추가합니다.
#### soup.select('<position>')은 해당 위치의 정보를 선택해 받습니다.
#### soup.select + .text는 해당 정보의 텍스트 부분받 추출합니다.
#### re. 패키지로 필요없는 부분을 제거할 수 있습니다.


### 평론가 의견 수집

#### 평론가들이 남긴 짧은 평을 수집합니다.

```python
for x, y, t in zip(gamelist_data['url'], gamelist_data['year'], gamelist_data['title']):
    url = basegamepageurl_critic % x 
    res = requests.get(url, headers = urlheader)
    soup = bsoup(res.content, 'html.parser')
    source = soup.select('div.source')
    score = soup.select('div.review_grade')
    body = soup.select('div.review_body')
    date = soup.select('div.date')
    for date, score, body, source in zip(date, score, body, source):
        bodysub = body.text
        bodysub= re.sub('\n                                ','',bodysub)
        bodysub = re.sub('\n                            ', '', bodysub)
        scorep = score.text
        scorep = re.sub('\n','',scorep)
        review_data_list.append([ t , y, date.text, scorep, bodysub, source.text])
```

```python
review_data_list
```

#### 데이터를 저장할 데이터 프레임 생성, 저장 정보는 타이틀,출시년도, 리뷰 날짜, 점수, 리뷰 내용, 출처

```python
review_data = pd.DataFrame(review_data_list, columns = ['TITLE','YEAR','DATE','SCORE','REVIEW','SOURCE'])
```

```python
review_data
```

```python
review_data.to_csv('Review_Data_Critic.csv')
```

### 유저 리뷰 수집

#### 게임을 즐긴 유저들이 남긴 리뷰를 수집합니다.

#### *유저 리뷰는 다른 사용자들이 유용하다 평가한 리뷰를 걸러서 추출합니다. 이는 유의미한 리뷰를 추출하고 쓸모없는 데이터를 추리기 위함입니다.

```python
user_review_data_list = []
```

```python
for x, y, t in zip(gamelist_data['url'], gamelist_data['year'], gamelist_data['title']):
    url = basegamepageurl_user % x 
    res = requests.get(url, headers = urlheader)
    soup = bsoup(res.content, 'html.parser')
    score = soup.select('div.review_grade')
    body = soup.select('div.review_body')
    date = soup.select('div.date')
    for date, score, body, source in zip(date, score, body, source):
        bodysub = body.text
        bodysub= re.sub('\n                                ','',bodysub)
        bodysub = re.sub('\n                            ', '', bodysub)
        bodysub = re.sub('\n', '', bodysub)
        scorep = score.text
        scorep = re.sub('\n','',scorep)
        user_review_data_list.append([ t , y, date.text, scorep, bodysub])
```

```python
user_review_data_list
```

#### 수집한 데이터를 데이터프레임화

```python
user_review_data = pd.DataFrame(user_review_data_list, columns = ['TITLE','YEAR','DATE','SCORE','REVIEW'])
```
