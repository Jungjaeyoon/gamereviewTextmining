#  Collect game review data from metacritic.com

## scrap top 100 game's review data from metacritic.com 

#### request 패키지를 통해 데이터를 받고 BeautifulSoup와 re로 처리합니다, 마지막으로 pandas를 통해 데이터 프레임 형태로 저장합니다.

```python
from bs4 import BeautifulSoup as bsoup
import requests 
import pandas as pd
import re
```

#### url 헤더는 각자 사용하는 환경에 맞춰 설정해주세요

urlheader = {'User-Agent' :'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:54.0) Gecko/20100101 Firefox/54.0' }

#### 수집하고 싶은 데이터를 받아오는 부분입니다.
#### 이번 분석에는 2015년 - 2017년 3년간 평가점수가 좋은 PS4 게임의 리뷰 정보를 수집하여 사용합니다.
#### 수집하고 싶은 페이지의 URL을 입력하고 requests.get 명령어를 통해 해당 페이지를 가져올 수 있습니다.

```python
gamelist_pageurl_2015 = requests.get('http://www.metacritic.com/browse/games/score/metascore/discussed/ps4/filtered?sort=desc&year_selected=2015', headers = urlheader)
gamelist_pageurl_2016 = requests.get('http://www.metacritic.com/browse/games/score/metascore/discussed/ps4/filtered?sort=desc&year_selected=2016', headers = urlheader)
gamelist_pageurl_2017 = requests.get('http://www.metacritic.com/browse/games/score/metascore/discussed/ps4/filtered?sort=desc&year_selected=2017', headers = urlheader)
gamelisturl = [(gamelist_pageurl_2015, 2015), (gamelist_pageurl_2016, 2016), (gamelist_pageurl_2017, 2017)]
```

#### gamelisturl 변수에 게임 목록과 url이 바로 저장되지는 않습니다.

```python
gamelisturl
```

#### request.get을 통해 수집한 정보는 beautifulsoup를 통해 읽어들여야 합니다.


```python
gamelist_list = []
for x in gamelisturl:
    soup = bsoup(x[0].text, 'html.parser')
    titles = soup.find_all('div', {'class': 'product_item product_title'})
    year = x[1]
    for i in titles:
        url = i.a['href']
        i = i.text
        i = re.sub('\n                                    \n','',i)
        i = re.sub('\n\n                    ', '' , i )
        gamelist_list.append([i,year,url])
gamelist_data = pd.DataFrame(gamelist_list, columns =['title', 'year','url'])
```

#### soup의 find_all 함수로 타이틀을 추출하고, 게임 리뷰 페이지의 url 형식이 http://www.metacritic.com/game/playstation-4 + /gamename 이므로 다음과 같은 형태로 저장했습니다.

```python
gamelist_data
```
